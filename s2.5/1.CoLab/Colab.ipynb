# Magic command
! 를 붙여서 bash 명령어들을 실행할 수 있다.

```
!ls
!pwd
```

```
!cd
!pwd
```

% 를 붙여 사용하면 해당내용이 반영된다. (cd 로 폴더를 옮길 때 유용)

```
%cd /
!pwd
```

```
%cd /content
```

참고사항.
- 코드라인 맨 위에 %%bash 를 적으면 bash명령어를 ! 를 적지 않고 실행할 수 있다.
